# Plot the function $2\cos(x)$ between $0$ and $pi$

```python
x_values = np.linspace(0, np.pi, 100)
x_values
```

```python
y_values = 2 * np.cos(x_values)
y_values
```

```python
plt.plot(x_values, y_values)
plt.xlabel('x val')
plt.ylabel('$2*\cos(x)$')
```
